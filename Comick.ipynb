```python
import traceback

# Kết nối đến cơ sở dữ liệu
conn = sqlite3.connect('manga.db')

# Tạo một đối tượng cursor để thực thi các truy vấn SQL
cursor = conn.cursor()

arr = []
arr2 = []
arr3=[]

for i in range(1,5):
    try:
        a = list_manga(i, 'hot', 'manhua')
        for j in range(len(a)):
            try:
                b = manga_info(a[j]['md_comics']['slug'])
                name = b['comic']['title']
                hid = b['comic']['hid']
                desc = b['comic']['desc']
                if len(b['authors'])==0: authors='Updating'
                else:authors = b['authors'][0]['name']
                covers = b['comic']['md_covers'][0]['b2key']
                slug = a[j]['md_comics']['slug']
                
                cursor.execute("SELECT COUNT(*) FROM manga_info WHERE hid = ?", (hid,))
                result = cursor.fetchone()
                if result[0] > 0:
                    # hid already exists in the database, skip inserting
                    continue

                insert_query = f"INSERT INTO manga_info (name, hid, desc, authors, covers, slug) VALUES (?, ?, ?, ?, ?, ?)"
                cursor.execute(insert_query, (name, hid, desc, authors, covers, slug))
                conn.commit()
            except Exception as e:
                traceback.print_exc()
                arr2.append(f"{a[j]['md_comics']['slug']}---{e}--{i}---{j}")
            
    except:
        arr.append(i)

# Đóng cursor và kết nối
cursor.close()
conn.close()
```
